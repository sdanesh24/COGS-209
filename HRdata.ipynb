```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
hrdata = pd.read_csv('batch2_jobID_00B80TR.csv')
```

```python
hrdata.head()
```

```python
hrdata.shape
```

```python
hrdata.describe()
```

```python
# EDA plots of jobType, degree, major, and industry against salary

sns.catplot(y = 'salary', x = 'jobType', data = hrdata, kind = 'boxen', height = 6, aspect = 3)
plt.show()
```

```python
# reporting median salaries of each jobType

mediansalary_job = hrdata.groupby(['jobType'])['salary'].median().reset_index()
mediansalary_job = mediansalary_job.sort_values(by='salary', ascending = False)

print(mediansalary_job)
```

```python
sns.catplot(y = 'salary', x = 'degree', data = hrdata, kind = 'boxen', height = 6, aspect = 3)
plt.show()
```

```python
# reporting median salaries of each degree

mediansalary_degree = hrdata.groupby(['degree'])['salary'].median().reset_index()
mediansalary_degree = mediansalary_degree.sort_values(by='salary', ascending = False)

print(mediansalary_degree)
```

```python
sns.catplot(y = 'salary', x = 'major', data = hrdata, kind = 'boxen', height = 6, aspect = 3)
plt.show()
```
